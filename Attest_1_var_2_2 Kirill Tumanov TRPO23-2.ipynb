1. Имеется список названий месяцев: [‘января’, ‘февраля’, ‘марта’, ‘апреля’, ’мая’, 
‘июня’, ‘июля’, ‘августа’, ‘сентября’, ‘октября’, ‘ноября’, ‘декабря’]. Создайте по этому 
списку словарь, в котором название месяца будет ключом, а номер месяца (от 1 до 12) –
значением. Используя полученный словарь преобразуйте строку с датой вида «1 января 
2021» в строку «1.01.2021»

```python
# Решение 1
# Создадим список месяцев
mes = ["января", "февраля", "марта", "апреля", "мая", "июня", "июля", "августа", "сентября", "октября", "ноября", "декабря"]
# Создадим словарь с номерами месяцев
mes_d = {}
for k, month in enumerate(mes):
    mes_d[month] = k+1
# Вводим строку с датой
s = input('Дата:')
# Делим по пробелам
s = s.split()

if mes_d[s[1]] < 10:
    res = f"{s[0]}.0{mes_d[s[1]]}.{s[2]}" #если число однозначное
else:
    res = f"{s[0]}.{mes_d[s[1]]}.{s[2]}" #если двузначное

print(res)
```

2. Имеется строка с названиями товаров вида «яблоки, груши, яблоки, киви, сливы, 
киви». Товары перечислены через запятую, товары могут повторяться. Выведите название 
товара, который встречается в этой строке чаще всего. Если таких товаров несколько, то 
названия всех товаров.

```python
string = "яблоки, груши, яблоки, киви, сливы, киви"

# Разделение строки на список товаров
products = string.split(", ")

# Создание словаря для подсчета количества товаров
col_t = {}

# Подсчет встречаемости каждого товара в словаре
for product in products:
    if product in col_t:
        col_t[product] += 1
    else:
        col_t[product] = 1

# Нахождение максимальной встречаемости
max_col = max(col_t.values())

# Вывод всех товаров, которые встречаются с максимальной частотой
maxt = [product for product, col in col_t.items() if col == max_col]

print(*maxt)
```

3. Имеется предложение, слова в котором, разделяются пробелами. С помощью списков получить слово, состоящее из первых букв слов в предложении.

```python
# Решение 3
#Вводим строку
s = input()
print(s)
# Делим её по пробелам и создаём список
a = s.split()

res = ''
# Прибавляем в результат первые буквы каждого слова
for x in a:
    res += x[0]

print(res)
```

4. Имеется переменная listlist, содержащая список списков. Заранее неизвестно, сколько в 
переменной списков и какой они длины (их длины в общем случае различны). Из этих списков 
составить список, содержащий поочередно элементы каждого из вложенных списков, при этом, 
порядок этих элементов сохраняется. Процедура создания списка останавливается, когда хотя 
бы в одном из исходных вложенных списков будут использованные все элементы. Кроме 
составленного списка получить список из всех элементов, не вошедших в составленный список. 
Пример: listlist =[[1,2,3],['a', 'b'], [30, 40, 50, 60]] -> [1, 'a', 30, 2, 'b', 40], [3, 50, 60]

```python
# Решение 4

listlist = [[1,2,3],['a', 'b'], [30, 40, 50, 60]]
# Находим минимальную длину вложенного списка
min_length = len(min(listlist, key=len))

x1 = []
x2 = []
# Cоставляем список, содержащий поочередно элементы каждого из вложенных списков, это x1, x2 - все остальные элементы
for i in range(0, min_length):
    for x in listlist:
        x1.append(x[i])

el = []
# Заносим все элементы в список el
for i in listlist:
    for x in i:
        el.append(x)
# Убираем элементы, которые попали в x1
x2 = list(set(el) - set(x1))
res = [x1,x2]
print(*res)
```
